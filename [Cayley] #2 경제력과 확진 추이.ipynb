```python
import plotly.express as px
import plotly as py
from plotly.offline import iplot
import plotly.graph_objects as go
```

```python
df = pd.read_csv(r"C:\Users\안지은\Desktop\Team.Cayley\owid-covid-data (1).csv")
df.head()
```

<h3> 전처리 : 홍콩 삭제 </h3>

```python
df.drop(df[df.location == 'Hong Kong'].index, 0,inplace = True)
df
```

<h3> G7국가 100만명당 누적확진자 수 </h3>

```python
df_eco = df[['date','location','total_cases_per_million','gdp_per_capita','extreme_poverty']]
```

```python
df_eco
```

```python
G7_cases = pd.DataFrame() 
G7 = ['United State', 'United Kingdom', 'France', 'Germany', 'Italy','Canada','Japan']
for loc in G7:
    G7_cases = G7_cases.append(df_eco[df_eco.location == loc]) 
```

```python
G7_cases
```

```python
fig = px.line(G7_cases,title = 'Total Cases per millon of G7 Member Countries',
              x="date", y="total_cases_per_million", color='location',hover_name = 'location')
fig
```

<h3>OECD국가 100만명당 누적확진자수</h3>

```python
OECD = ['Austria', 'Australia', 'Belgium', 'Canada', 'Chile', 'Colombia', 'Czech Republic', 'Denmark', 'Estonia', 'Finland', 'France', 'Germany', 'Greece', 'Hungary','Iceland','Ireland', 'Israel', 'Italy', 'Japan', 'South Korea', 'Latvia', 'Lithuania', 'Luxembourg', 'Mexico', 'Netherland', 'New Zealand', 'Norway', 'Poland', 'Portugal','Slovak Republic', 'Slovenia', 'Spain', 'Sweden', 'Switzerland', 'Turkey', 'United Kingdom', 'United States']
OECD_cases = pd.DataFrame() 

for loc in OECD:
    OECD_cases = OECD_cases.append(df_eco[df_eco.location == loc]) 
```

```python
OECD_cases
```

```python
fig = px.line(OECD_cases, title = 'Total Cases per millon of OECD Member Countries',
              x="date", y="total_cases_per_million", color='location',hover_name = 'location')
fig
```

<h3> GDP하위 10위 국가 100만명당 누적확진자수 </h3>

```python
gdp_low = pd.DataFrame(df.groupby('location')['gdp_per_capita'].mean().sort_values().head(10)).reset_index()
gdp_low
```

```python
gdp_tail_10 = gdp_low.location.tolist()
```

```python

tail_cases = pd.DataFrame() 

for loc in gdp_tail_10:
    tail_cases = tail_cases.append(df_eco[df_eco.location == loc]) 
```

```python
gdp_tail_10[:]
```

```python
fig = px.line(tail_cases, title = 'Total Cases per millon of The bottom 10 countries of GDP', x="date",
              y="total_cases_per_million",hover_name = 'location',color = 'location')
fig
```
